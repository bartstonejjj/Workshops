Fix column headers

```python
df = data.copy()
rounds = list(range(1,25))
df.columns = ['team'] + rounds
data2 = df
df.head()
```

Add year column

```python
df = data2.copy()
df['year'] = df['team'].str.extract('(\d{4})')
data3 = df
df.head()
```

Remove extraneous columns and rows

```python
df = data3.copy()
df['year_row'] = df['year'].notnull() # label 'year rows' (which are extraneous)
df['year'] = df['year'].ffill() # Fill year rows down to the bottom
df = df[~df['year_row']].drop(['year_row'], axis = 1) # Remove rows that are a 'year row' and remove column 'year_row'
df = df[~df['team'].isin(['Team', '[Top]'])] # Remove 'Team' and '[Top]' rows
data4 = df
df.head()
```

Check there are no more extraneous rows or columns

```python
df = data4.copy()
df['team'].value_counts()
```

Remove extraneous asterisk in team name

```python
df = data4.copy()
df['team'] = df['team'].map(lambda x: x[1:] if x[0] == '*' else x)
data5 = df
df['team'].value_counts()
```

Check if 'data' columns are numeric

```python
data5.info()
```

```python
df = data5.copy()
for r in [1]:
    print(df[r].value_counts())
```

Remove extraneous characters from 'data' columns

```python
df = data5.copy()
df[1] = df[1].str.extract('(\d+)')
df[1] = df[1].astype(float)
data6 = df
df[1].value_counts()
```

Show round by round ladder positions for a given team for each season (to validate data with human interaction) to see if appears as expected

```python
df = data6.copy()
team = 'Adelaide'
df = df[df['team'] == team]
for year in df['year']:
    tmp = df[df['year'] == year][rounds].transpose().dropna()
    tmp = tmp[tmp.columns[0]]
    tmp.plot(title = 'Ladder positions round by round for {} in season: {}.'.format(team, year), ylim = (0, 18), figsize = (20, 10))
    plt.show()
```
